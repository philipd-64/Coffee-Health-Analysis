# Import

# Sleep Quality

```python
df['Sleep_Quality'].value_counts()
```

```python
df['Sleep_Quality'].replace('Poor', '1')
```

```python
df['Sleep_Quality'] = df['Sleep_Quality'].replace('Poor', 1)
```

```python
df['Sleep_Quality'] = df['Sleep_Quality'].replace('Fair', 2)
```

```python
df['Sleep_Quality'] = df['Sleep_Quality'].replace('Good', 3)
```

```python
df['Sleep_Quality'] = df['Sleep_Quality'].replace('Excellent', 4)
```

```python
df['Sleep_Quality'].value_counts()
```

```python
df.dtypes
```

# Stress Level

```python
df['Stress_Level'].value_counts()
```

```python
df['Stress_Level'] = df['Stress_Level'].replace('Low', 1)
df['Stress_Level'] = df['Stress_Level'].replace('Medium', 2)
df['Stress_Level'] = df['Stress_Level'].replace('High', 3)
```

```python
df.dtypes
```

# Health Issues

```python
df['Health_Issues'].value_counts()
```

```python
df['Health_Issues'] = df['Health_Issues'].replace('Zero', 1)
df['Health_Issues'] = df['Health_Issues'].replace('Mild', 2)
df['Health_Issues'] = df['Health_Issues'].replace('Moderate', 3)
df['Health_Issues'] = df['Health_Issues'].replace('Severe', 4)
```

```python
df['Health_Issues'].value_counts()
```
